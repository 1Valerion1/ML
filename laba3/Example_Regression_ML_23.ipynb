```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = df.rename(columns={'Cement (component 1)(kg in a m^3 mixture)':'Cement',
                        'Blast Furnace Slag (component 2)(kg in a m^3 mixture)':'Slag',
                        'Fly Ash (component 3)(kg in a m^3 mixture)':'Ash',
                        'Water  (component 4)(kg in a m^3 mixture)':'Water',
                        'Superplasticizer (component 5)(kg in a m^3 mixture)':'Superplasticizer',
                        'Coarse Aggregate  (component 6)(kg in a m^3 mixture)':'Coarse Aggregate',
                        'Fine Aggregate (component 7)(kg in a m^3 mixture)':'Fine Aggregate',
                        'Age (day)':'Age',
                        'Concrete compressive strength(MPa, megapascals) ':'Strength'
                        })
```

```python
import seaborn as sns
mask = np.triu(df.corr())

ax = sns.heatmap(df.corr(), cmap="vlag", annot=True, fmt="0.2f", mask = mask)
for t in ax.texts:
    if float(t.get_text()) >= 0.3 or float(t.get_text()) <= -0.3:
        t.set_text(t.get_text())
    else:
        t.set_text("")
```

```python
df.columns
```

```python
X=df.drop(['Strength'],axis=1)
y=df['Strength']
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
X_scaled = min_max_scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,
                                                 test_size=0.2,random_state=1)
print('Initial size = ',X_scaled.shape,' Train size = ',X_train.shape,
      'Test size = ', X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train,y_train)
```

```python
print('weight_0 =',np.round(linear_model.intercept_,3))
print('weights = ',np.round(linear_model.coef_,3))
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
y_pred_linear_model = linear_model.predict(X_test)
print('Linear Regression Estimation:')
print('R2-score:',round(r2_score(y_test,y_pred_linear_model),3))
print('MSE:',round(mean_squared_error(y_test,y_pred_linear_model),3))
print('MAE:',round(mean_absolute_error(y_test,y_pred_linear_model),3))
```

```python
from sklearn import linear_model
lasso_model = linear_model.Lasso(alpha=0.2)
lasso_model.fit(X_train, y_train)

print('weight_0 =',np.round(lasso_model.intercept_,3))
print('weights = ',np.round(lasso_model.coef_,3))
```

```python
y_pred_lasso_model = lasso_model.predict(X_test)
print('Lasso Regression Estimation:')
print('R2-score:',round(r2_score(y_test,y_pred_lasso_model),3))
print('MSE:',round(mean_squared_error(y_test,y_pred_lasso_model),3))
print('MAE:',round(mean_absolute_error(y_test,y_pred_lasso_model),3))
```

```python
from sklearn import linear_model
ridge_model = linear_model.Ridge(alpha=1)
ridge_model.fit(X_train, y_train)

print('weight_0 =',np.round(ridge_model.intercept_,3))
print('weights = ',np.round(ridge_model.coef_,3))
```

```python
y_pred_ridge_model = ridge_model.predict(X_test)
print('Ridge Regression Estimation:')
print('R2-score:',round(r2_score(y_test,y_pred_ridge_model),3))
print('MSE:',round(mean_squared_error(y_test,y_pred_ridge_model),3))
print('MAE:',round(mean_absolute_error(y_test,y_pred_ridge_model),3))
```

```python
from sklearn.linear_model import ElasticNet

elastic_net_model = ElasticNet(alpha=0.1, l1_ratio = 0.8)
elastic_net_model.fit(X_train, y_train)
print('weight_0 =',np.round(elastic_net_model.intercept_,3))
print('weights = ',np.round(elastic_net_model.coef_,3))

y_pred_elastic_net_model = elastic_net_model.predict(X_test)
print('Elastic Net Regression Estimation:')
print('R2-score:',round(r2_score(y_test,y_pred_elastic_net_model),3))
print('MSE:',round(mean_squared_error(y_test,y_pred_elastic_net_model),3))
print('MAE:',round(mean_absolute_error(y_test,y_pred_elastic_net_model),3))
```

```python
from sklearn.linear_model import ElasticNetCV

lin_regressor = ElasticNetCV(l1_ratio = 0.4, n_alphas=200, cv=10, random_state = 1)
lin_regressor.fit(X_train, y_train)

print('weight_0 =',np.round(lin_regressor.intercept_,3))
print('weights = ',np.round(lin_regressor.coef_,3))

y_pred_lin_reg_model = lin_regressor.predict(X_test)

print(f'Alpha value (penalty) = {round(lin_regressor.alpha_,4)}')

print('Elastic Net Regression Estimation:')
print('R2-score:',round(r2_score(y_test,y_pred_lin_reg_model),3))
print('MSE:',round(mean_squared_error(y_test,y_pred_lin_reg_model),3))
print('MAE:',round(mean_absolute_error(y_test,y_pred_lin_reg_model),3))
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline(steps=[
    ('scaling', MinMaxScaler()),
    ('polynomial_features', PolynomialFeatures(degree=3)),
    ('linear_model', LinearRegression())])

X = df.drop(['Strength'],axis=1)
y = df['Strength']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
pipeline.fit(X_train, y_train)

y_pred_pipeline = pipeline.predict(X_test)
print('Pipeline Performance:')
print('R2-score:',round(r2_score(y_test,y_pred_pipeline),3))
print('MSE:',round(mean_squared_error(y_test,y_pred_pipeline),3))
print('MAE:',round(mean_absolute_error(y_test,y_pred_pipeline),3))
```
